   Bayes' theorem:                                                                    
       P(A|B) = Posterior                                                             
       P(B) = Marginal
       P(B|A) = Likelihood
       P(A) = Prior
       
P(D+|T+) = P(T+|D+)* P(D+)/P(T+|D+)* P(D+)+P(T+|D-)* P(D-) 
or 
P(A|B)= P(B|A)P(A)/P(B)

# Expressions for Baye's theorem
Sample =1, Disease present P(D+) or No Disease P(D-)
P(D+) w/ Positive Test P(T+)= True Positive P(T+|D+) or P(D+) w/ Negative Test P(T-)= False Negative P(T-|D+)
(D-) w/ Positive Test P(T+)= False Positive  P(T+|D-)or P(D-) w/ Negative Test (T-) True Negative  P(T-|D-)   
P(D+)+P(D-)= 1, P(T+|D+)+P(T-|D+)=1,P(T+|D-)+P(T-|D-)=1
Ex:
P(D+)(0.20): P(T+|D+)(0.974)+ P(T-|D+)(0.026)=1
P(D-)(0.80):P(T+|D-)(0.051)+ P(T-|D-)(0.949) = 1
Form: P(D+|T+)=P(T+|D+)P(D+)/P(T+|D+)xP(D+)+P(T+|D-)xP(D-)=(0.974x0.20)+(0.051x0.80)/(0.974x0.20)
